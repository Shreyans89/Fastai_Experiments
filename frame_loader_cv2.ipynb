```python
import cv2
import numpy as np



img_folder='C:/Users/ANANYA/Desktop/FASTAI_WD_LATEST/cv2_imgs'
vidcap = cv2.VideoCapture('VIRAT_S_000203_07_001341_001458.mp4')
success,image = vidcap.read()
count = 0
while success:
    cv2.imwrite(img_folder+'/'+str(count)+'.jpg', image)       # save frame as JPEG file      
    success,image = vidcap.read()
    print('Read a new frame: ', success)
    count += 1
    

```
